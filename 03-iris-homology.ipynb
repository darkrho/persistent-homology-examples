```python
%load_ext rmagic
```

```python
%%R
data <- as.matrix(iris[,-5])
head(data)
summary(data)
```

```python
from IPython.display import Image
Image("files/Iris_data_set.png")
```

```python
%%R -w 12 -h 12 -u in
library(phom)
max_dim <- 2
max_f <- 1
intervals <- pHom(data, max_dim, max_f, mode="vr", metric="euclidean")
par(mfrow=c(2,2))
plotBarcodeDiagram(intervals, 0, max_f, title="H0 Barcode of Iris Data")
plotBarcodeDiagram(intervals, 1, max_f, title="H1 Barcode of Iris Data")
plotBarcodeDiagram(intervals, 2, max_f, title="H2 Barcode of Iris Data")
plotPersistenceDiagram(intervals, max_dim, max_f, title="Persistence Diagram of Iris Data")
```
