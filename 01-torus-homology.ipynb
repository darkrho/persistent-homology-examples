# The Torus Homology

```python
%load_ext rmagic
```

```python
%%R
library(phom)
library(scatterplot3d)

N <- 20
R <- 25
r <- 10
sd <- 1

ts <- seq(0,2*pi,pi/N)
ps <- seq(0,2*pi,pi/N)

x<-y<-z<-c()
for (t in ts) for (p in ps) {
  x <- c(x,R*cos(t)+r*cos(p)*cos(t))
  y <- c(y,R*sin(t)+r*cos(p)*sin(t))
  z <- c(z,r*sin(p))
}
```

```python
%%R -w 12 -h 6 -u in
par(mfcol=c(1,2))
scatterplot3d(x,y,z,type="l")
```

```python
%%R -w 12 -h 8 -u in
par(mfrow=c(2,2))
max_dim <- 2
max_f <- 3.9
intervals <- pHom(cbind(x,y,z), max_dim, max_f, mode="lw")
plotBarcodeDiagram(intervals, 0, max_f, title="H0")
plotBarcodeDiagram(intervals, 1, max_f, title="H1")
plotBarcodeDiagram(intervals, 2, max_f, title="H2")
plotPersistenceDiagram(intervals, max_dim, max_f)
```
