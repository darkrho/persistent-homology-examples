# Anscombe's Quartet Homology

```python
%matplotlib inline
import matplotlib.pylab as plt
import seaborn as sns
figsize = (16, 12)
sns.set_context("talk", font_scale=1., rc={"figure.figsize": figsize})
sns.set_style("ticks")
plt.rcParams["figure.figsize"] = figsize
```

```python
%load_ext rmagic
import pandas.rpy.common as rcom
```

```python
df = sns.load_dataset("anscombe")
df.groupby("dataset").describe().unstack()
```

```python
g = sns.lmplot("x", "y", col="dataset", hue="dataset", data=df,
               col_wrap=2, ci=None, palette="muted", size=4,
               scatter_kws={"s": 50, "alpha": 1}, legend=False)
plt.tight_layout()
```

```python
df1 = df[df.dataset=="I"][["x", "y"]]
df2 = df[df.dataset=="II"][["x", "y"]]
df3 = df[df.dataset=="III"][["x", "y"]]
df4 = df[df.dataset=="IV"][["x", "y"]]
%Rpush df1 df2 df3 df4
```

```python
%%R
library(phom)
max_dim <- 1
max_f <- 5
int1 <- pHom(as.matrix(df1), max_dim, max_f)
int2 <- pHom(as.matrix(df2), max_dim, max_f)
int3 <- pHom(as.matrix(df3), max_dim, max_f)
int4 <- pHom(as.matrix(df4), max_dim, max_f)
```

```python
%%R -w 12 -h 6 -u in
par(mfrow=c(2, 2))
plotBarcodeDiagram(int1, 0, max_f, title="H0 - Dataset I")
plotBarcodeDiagram(int2, 0, max_f, title="H0 - Dataset II")
plotBarcodeDiagram(int3, 0, max_f, title="H0 - Dataset III")
plotBarcodeDiagram(int4, 0, max_f, title="H0 - Dataset IV")
```

```python
%%R -w 12 -h 6 -u in
par(mfrow=c(2, 2))
plotBarcodeDiagram(int1, 1, max_f, title="H1 - Dataset I")
plotBarcodeDiagram(int2, 1, max_f, title="H1 - Dataset II")
plotBarcodeDiagram(int3, 1, max_f, title="H1 - Dataset III")
plotBarcodeDiagram(int4, 1, max_f, title="H1 - Dataset IV")
```
